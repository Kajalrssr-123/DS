```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data for the plots
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a figure with multiple subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Subplot 1 (Upper Left)
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('Sin(x)')

# Subplot 2 (Upper Right)
axs[0, 1].plot(x, y2, color='red')
axs[0, 1].set_title('Cos(x)')

# Subplot 3 (Lower Left)
axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('Tan(x)')

# Hide the empty subplot (Lower Right)
axs[1, 1].axis('off')

# Adjust spacing between subplots
plt.tight_layout()

# Set a common title for all subplots
fig.suptitle('Multiple Plots')

# Display the subplots
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
languages = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]

# Create a bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(languages, popularity, color='skyblue')

# Add labels and a title
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the bar chart
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
languages = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]

# Generate a list of different colors for each bar
colors = ['red', 'green', 'blue', 'orange', 'purple', 'pink']

# Create a bar chart with different colors
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
bars = plt.bar(languages, popularity, color=colors)

# Add labels and a title
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')

# Custom legend for the colors
legend_labels = [plt.Line2D([0], [0], color=color, lw=4, label=language) for language, color in zip(languages, colors)]
plt.legend(handles=legend_labels, title="Languages")

# Display the bar chart
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
men_means = [22, 30, 35, 35, 26]
women_means = [25, 32, 30, 35, 29]

# Set the width of the bars
bar_width = 0.35

# Generate X values for Men and Women
x_men = np.arange(len(groups))
x_women = x_men + bar_width

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(x_men, men_means, width=bar_width, label='Men', color='g', align='center')
plt.bar(x_women, women_means, width=bar_width, label='Women', color='r', align='edge')

# Add labels and a title
plt.xlabel('Groups')
plt.ylabel('Means')
plt.title('Scores by Group and Gender')
plt.xticks(x_men + bar_width / 2, groups)

# Add a legend
plt.legend()

# Display the bar plot
plt.tight_layout()
plt.show()
```
